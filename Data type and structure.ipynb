**Q1. What are data structures, and why are they important ?**

**Q2. Explain the difference between mutable and immutable data types with examples ?**

**Explanation** :
> Mutable: Can be changed after creation.
>> Example:

```python
my_list = [1, 2, 3]
my_list[0] = 10  # list is mutable
print(my_list)
```

Immutable: Cannot be changed after creation.
>Example:




```python
my_string = "hello"
# my_string[0] = "H" is not allowed because if we do that this will generate error
print(my_string)
```

**Q3. What are the main differences between lists and tuples in Python ?**

** Explanation :**
 1. Syntax :

  * List: Written with square brackets []
  * my_list = [1, 2, 3]

  * Tuple: Written with parentheses ()
  * my_tuple = (1, 2, 3)

2. Mutability :

 * List: Mutable -> You can add, remove, or change elements.
 * my_list[0] = 10   # It will give output as 10 at index 0

  * Tuple: Immutable -> Once created, it cannot be changed.

  * my_tuple[0] = 10   # error

3. Performance

  * List: Slightly slower (because it supports modifications).

  * Tuple: Faster (because it's fixed and cannot change).

4. Use Cases

  * List: Used when you need a dynamic collection (adding/removing items).

  * Tuple: Used when you need fixed data (like days of the week, coordinates).

5. Methods Available

  * List: Has many methods like append(), remove(), sort(), etc.

  * Tuple: Has only two methods -> count() and index().
